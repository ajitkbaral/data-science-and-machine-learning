```python
import numpy as np
x = np.arange(100)
y = x * 2
z = x ** 2
```

```python
# ** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. 
# What command do you use if you aren't using the jupyter notebook?**
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
'''
Exercise 1
** Follow along with these steps: **

** Create a figure object called fig using plt.figure() **
** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
** Plot (x,y) on that axes and set the labels and titles to match the plot below:**
'''
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('title')
axes.set_xlabel('x')
axes.set_ylabel('y')
```

```python
'''
Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. 
Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**
'''
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
# ** Now plot (x,y) on both axes. And call your figure object to show it.**
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.plot(x,y, color='brown')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.plot(x,y, color='brown')
```

```python
# ** Now use x,y, and z arrays to recreate the plot below. 
# Notice the xlimits and y limits on the inserted plot:**
```

```python
fig = plt.figure()
outer = fig.add_axes([0,0,1,1])
outer.set_xlabel('X')
outer.set_ylabel('Z')

inner = fig.add_axes([0.2,0.5,0.4,0.4])
inner.set_xlabel('X')
inner.set_ylabel('Y')
inner.set_title('zoom')
inner.set_xlim([20, 22])
inner.set_ylim([30, 50])

outer.plot(x, z)
inner.plot(x, y)
```

```python
fig = plt.subplots(nrows=1, ncols=2)
```

```python
# ** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x, y, color='blue', lw=3, ls='--')
axes[1].plot(x, z, color='red', lw=3)
```

```python
# ** See if you can resize the plot by adding the figsize() argument in plt.subplots() 
# are copying and pasting your previous code.**
```

```python
fig, axes = plt.subplots(1,2, figsize=(12, 2))
axes[0].plot(x, y, color='blue', lw=5)
axes[1].plot(x, z, color='red', lw=2, ls='--')
```
