```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# Functional method
```

```python
plt.plot(x,y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Left, Bottom, Width, Height
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGET PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current in axes:
    current.plot(x,y)
    
axes[0].plot(y,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# Figure Size, DPI and Legend
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_figure.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title("Title")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title("Title")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.legend(loc=1)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title("Title")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.legend(loc=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title("Title")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.legend(loc=(0.1, 0.2))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color="orange")
axes.plot(x,y, color="green")
axes.plot(x,y, color="#FF8C00")
axes.plot(x,y, color="purple", linewidth=20, alpha=0.5)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color="orange")
axes.plot(x,y, color="green")
axes.plot(x,y, color="#FF8C00")
axes.plot(x,y, color="purple", lw=3, linestyle='--')
axes.plot(x,y, color="purple", lw=3, ls='--', drawstyle='steps')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# axes.plot(x,y, color="purple", lw=3, ls='--', marker='o', markersize=20)
axes.plot(x,y, color="purple", lw=3, ls='--', marker='+', markersize=20, markerfacecolor='green')
```
